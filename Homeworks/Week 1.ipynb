### Question 7

```python
# Selecting Lotus cars 

df_lotus = df.loc[df['Make'] == 'Lotus']
df_lotus
```

```python
# Selecting columns "Engine HP" and "Engine Cylinders"

result_df = df_lotus[["Engine HP", "Engine Cylinders"]]
result_df
```

```python
# Dropping duplicates

df1 = result_df.drop_duplicates()
df1
```

```python
# Getting the underlying array from result_df

x = df1.to_numpy()
x
```

```python
# Matrix-matrix multiplication between transpose of x and x

x_transpose = x.T

XTX = x_transpose.dot(x)
XTX
```

```python
# Inverting XTX

XTX_inverse = np.linalg.inv(XTX)
print(f'XTX inverted:\n{XTX_inverse}')
```

```python
# creating array y
y = np.array([1100, 800, 750, 850, 1300, 1000, 1000, 1300, 800])
y
```

```python
# Transposing x
x_transpose = x.T

w = (XTX_inverse.dot(x_transpose)).dot(y)
w
```
